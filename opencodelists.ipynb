```python
import pandas as pd

table = soup.find_all('table')
df = pd.read_html(str(table))[0]
```

```python
df
```

```python
import json

valueSetJson = {
    "resourceType": "ValueSet",
    "status": "active",
}
print(json.dumps(valueSetJson, indent=2))
```

```python
from fhir.resources.valueset import ValueSet
valueSet = ValueSet(**valueSetJson)

print(valueSet)
```

```python
valueSetJson['compose'] = { 'include': [{ "system" : "http://snomed.info/sct","concept" : []}]}
print(json.dumps(valueSetJson, indent=2))
```

```python
for index, row in df.iterrows():
    valueSetJson['compose']['include'][0]['concept'].append({
        "code" : str(row[0]),
        "display" : row[1]
    })
print(json.dumps(valueSetJson, indent=2))
```

```python
valueSet = ValueSet(**valueSetJson)

print(valueSet)
```
