```python

import pandas as pd
import requests

headers = {'User-Agent': 'Mozilla/5.0 (X11; Windows; Windows x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.5060.114 Safari/537.36'}

url = 'https://www.opencodelists.org/codelist/nhsd-primary-care-domain-refsets/c19haemcan_cod/20241205/#full-list'
s = requests.session()
r = s.get(url)
```

```python
if r.ok:
    print(r.text)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text, "lxml")
table = soup.find_all('div', 'list-group-item py-1 px-2')
table[0]
```

```python
import json

valueSetJson = {
    "resourceType": "ValueSet",
    "status": "active",
}
print(json.dumps(valueSetJson, indent=2))
```

```python
df = pd.read_xml(str(table[3]))
id = df.loc[df.index[1], 'dd']
id = id.replace('/','-').replace('_','-')
valueSetJson['id']=id
valueSetJson['url'] = "https://fhir.opencodelists.org/ValueSet/"+id

df = pd.read_xml(str(table[4]))
version = str(df.loc[df.index[1], 'dd'])
valueSetJson['version']=version

valueSetJson['date'] = version[0:4] + "-" +version[4:6] + "-" +version[6:8]

df = pd.read_xml(str(table[2]))
publisher = str(df.loc[df.index[1], 'dd'])
valueSetJson['publisher'] = publisher
#print(json.dumps(valueSetJson, indent=2))
```

```python
from fhir.resources.valueset import ValueSet

table = soup.find_all('h1','h3')
mystring = table[0].text
mystring = mystring.replace('\n', ' ').replace('\r', '').strip()

valueSetJson['title']=mystring
print(json.dumps(valueSetJson, indent=2))
valueSet = ValueSet(**valueSetJson)
```

```python
valueSetJson['compose'] = { 'include': [{ "system" : "http://snomed.info/sct","concept" : []}]}
```

```python
for index, row in df.iterrows():
    display = str(row[1])
    valueSetJson['compose']['include'][0]['concept'].append({
        "code" : str(row[0]),
        "display" : display
    })
print(json.dumps(valueSetJson, indent=2, ensure_ascii=False))
```

```python
valueSet = ValueSet(**valueSetJson)

print(valueSet)
```
