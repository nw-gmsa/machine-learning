# PyPDF2

```python
from PyPDF2 import PdfFileReader, PdfReader

#pdf_file = "PDF/Genomics.pdf"
pdf_file = "PDF/genome-report-example.pdf"
pdf_object = PdfReader(pdf_file)
```

```python
page_count = len(pdf_object.pages)
# Access a specific page (index starts at 0)
page = pdf_object.pages[0]
```

## text

```python
all_text = ""
for page in pdf_object.pages:
    all_text += page.extract_text() + "\n"
print(all_text)
```

```python

information = pdf_object.metadata
number_of_pages = len(pdf_object.pages)
txt = f"""
Author: {information.author}
Creator: {information.creator}
Producer: {information.producer}
Subject: {information.subject}
Title: {information.title}
Number of pages: {number_of_pages}
"""
# Here the metadata of your pdf
print(txt)
```

# pymupdf


```python
import pymupdf # imports the pymupdf library

pdf_file = "PDF/genome-report-example.pdf"
doc = pymupdf.open(pdf_file)
```

## text

```python
for page in doc:  # iterate the document pages
    page.read_contents()
    text = page.get_text()
    print(text)
```

## Html

```python

for page in doc:  # iterate the document pages
    page.read_contents()
    text = page.get_text('html', clip = None).encode("utf8")
    print(text)  # write text of page
with open("demofile.html", "wb", ) as f:
    for page in doc:  # iterate the document pages
        page.read_contents()
        text = page.get_text('html', clip = None).encode("utf8")
        f.write(text)
```
